```python
%matplotlib inline
import sys
sys.path.append('../../')

from bitc.experiments import Experiment
from bitc.models import Calibration
from bitc.instruments import Instrument
import pymc
import numpy
import seaborn

experiment_name = 'water_titration'
datafile = 'water_water.itc'

experiment = Experiment(datafile, experiment_name)
instrument = Instrument(itcfile=datafile)
model = Calibration(experiment, instrument)

print("Fitting")
map = pymc.MAP(model)
map.fit(iterlim=2000)

print("Sampling...")
model.mcmc.sample(iter=20000, burn=250, thin=25)

pymc.Matplot.plot(model.mcmc.trace('DeltaH_0')[:] , '%s-DeltaH_0' % experiment_name)
pymc.Matplot.plot(numpy.exp(model.mcmc.trace('log_sigma')[:]), '%s-sigma' % experiment_name)

model.mcmc.summary?



```
